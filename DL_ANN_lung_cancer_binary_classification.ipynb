```python

import pandas as pd
df = pd.read_csv('lung_cancer_examples.csv')
df
```

```python
X = df.iloc[:,2:6].values    # or can drop name n surname
X
```

```python
y = df.iloc[:,-1].values
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
X_scaled = scalar.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y,test_size=0.3,random_state=1)
```

```python
from keras.models import Sequential
from keras.layers import Dense   # dense means fully connected layer
ann = Sequential()
```

```python
# input layer
ann.add(Dense(4,activation='relu'))    # units =4 bcoz 4 features. relu gives better o/p

# 1st hidden layer
ann.add(Dense(6,activation='relu'))    # units =4

# 2nd hidden layer
ann.add(Dense(10,activation='relu'))    # units =4

# o/p layer . for binary classification dataset, 1 neuron is needed. we use sigmoid activation function

ann.add(Dense(1,activation='sigmoid'))    # for o/p layer either sigmoid/softmax is used. since this is a binary classification pblm, sigmoid is used
#
```

```python
# compile model

ann.compile(optimizer='adam', loss="binary_crossentropy", metrics=['accuracy'])    # adaptivie momentum optimizer(adam) is for updating weight , loss= diff of actual n predicted value, specify which metric performance to be improved by updating weight
```

```python
ann.fit(X_train,y_train, epochs=100, batch_size=15)  # epoch means how many times model shud learn, batch size is to split data for learning as batches,eg: 150samples batch-size=50 then 1 batch contain 50 samples,so total 3 batches needed for 150samples

```

```python
# find no: of edges
ann.summary()    # 4 is no: of i/p, 5 is no: of neurons(4 + 1 bias). 4*5 =20
```

Apply on test data

```python
loss, accuracy = ann.evaluate(X_test,y_test)
print('loss ',loss)
print('accuracy ',accuracy)
```

```python
# predict
# when using sigmoid we get a value less than 1 like .87 or .42 or .71 etc. so to predict whether i falls to class 0 or 1 ,compare it with 0.5

y_new = ann.predict(scalar.transform([[35,7,5,4]]))
if y_new.item() >=0.5:
  print("suffer lung cancer")
else:
  print("no cancer")
```
